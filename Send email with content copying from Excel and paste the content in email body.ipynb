```python
import openpyxl
import win32com.client as win32

def read_excel_and_send_email(excel_file_path, email_address):
    # Load the Excel workbook
    workbook = openpyxl.load_workbook(excel_file_path)
    sheet = workbook['Sheet1']

    # Prepare the email body
    email_body = "Data from Excel Sheet1:\n\n"
    for row in sheet.iter_rows(min_row=2, max_col=3, values_only=True):
        if row[0] is not None and row[0] != 0:
            email_body += f"{row[0]} | {row[1]} | {row[2]}\n"

    # If no valid data found, return
    if email_body == "Data from Excel Sheet1:\n\n":
        print("No valid data found to send email.")
        return

    # Set up the Outlook application
    outlook = win32.Dispatch('outlook.application')
    mail = outlook.CreateItem(0)
    mail.To = email_address
    mail.Subject = "Data from Excel Sheet1"
    mail.Body = email_body

    # Send the email
    mail.Send()

    print("Email sent successfully.")

# Example usage
excel_file_path = "C:/Users/phong/OneDrive/Desktop/auto_email/auto_email.xlsx"
email_address = "phongnthai@gmail.com"
read_excel_and_send_email(excel_file_path, email_address)
```

```python
import openpyxl
import win32com.client as win32
from openpyxl.styles import PatternFill

def read_excel_and_send_email(excel_file_path, email_address):
    # Load the Excel workbook
    workbook = openpyxl.load_workbook(excel_file_path)
    sheet = workbook['Sheet1']

    # Prepare the email body
    email_body = "<html><body>Data from Excel Sheet1:<br><br><table border='1'>"
    
    # Copy header row
    email_body += "<tr>"
    for cell in sheet.iter_rows(min_row=1, max_row=1, values_only=True):
        for header_cell in cell:
            email_body += f"<th>{header_cell}</th>"
    email_body += "</tr>"

    for row in sheet.iter_rows(min_row=2, max_col=3):
        if row[0].value is not None and row[0].value != 0:
            email_body += "<tr>"
            for cell in row:
                if isinstance(cell.value, (int, float)):
                    email_body += f"<td>{cell.value}</td>"
                else:
                    email_body += f"<td>{cell.value}</td>"
            email_body += "</tr>"

    email_body += "</table></body></html>"

    # If no valid data found, return
    if email_body == "<html><body>Data from Excel Sheet1:<br><br><table border='1'></table></body></html>":
        print("No valid data found to send email.")
        return

    # Set up the Outlook application
    outlook = win32.Dispatch('outlook.application')
    mail = outlook.CreateItem(0)
    mail.To = email_address
    mail.Subject = "Data from Excel Sheet1"
    mail.HTMLBody = email_body

    # Send the email
    mail.Send()

    print("Email sent successfully.")

# Example usage
excel_file_path = "C:/Users/phong/OneDrive/Desktop/auto_email/auto_email.xlsx"
email_address = "phongnthai@gmail.com"
read_excel_and_send_email(excel_file_path, email_address)
```

```python
import win32com.client as win32
import openpyxl
import os

def read_excel_and_send_email(excel_file_path, email_address):
    # Load the Excel workbook
    wb = openpyxl.load_workbook(excel_file_path)
    ws = wb['Sheet1']

    # Initialize variables to store table data
    table_data = []

    # Iterate through rows and columns to extract table data
    for row in ws.iter_rows(min_row=1, max_row=50, max_col=3, values_only=True):
        if row[0] not in (None, 0):
            table_data.append(row)

    # Close the workbook
    wb.close()

    # Create a new workbook to store filtered data
    filtered_wb = openpyxl.Workbook()
    filtered_ws = filtered_wb.active

    # Write filtered data to the new workbook
    for row in table_data:
        filtered_ws.append(row)

    # Copy conditional formatting from original sheet to the new sheet
    for row in ws.iter_rows(min_row=1, max_row=50, max_col=3):
        for cell in row:
            new_cell = filtered_ws[cell.coordinate]
            new_cell._style = cell._style

    # Save the filtered workbook
    filtered_excel_file_path = os.path.join(os.path.dirname(excel_file_path), "filtered_data.xlsx")
    filtered_wb.save(filtered_excel_file_path)

    # Send email with filtered data as attachment
    outlook = win32.Dispatch('outlook.application')
    mail = outlook.CreateItem(0)
    mail.To = email_address
    mail.Subject = "Data from Excel Sheet1"

    # Add the filtered data Excel file as an attachment
    if os.path.exists(filtered_excel_file_path):
        mail.Attachments.Add(filtered_excel_file_path)
    else:
        print("Filtered Excel file not found at the specified path.")

    # Send the email
    mail.Send()

    print("Email sent successfully.")

    # Clean up: delete the filtered Excel file
    os.remove(filtered_excel_file_path)

# Example usage
excel_file_path = "C:/Users/phong/OneDrive/Desktop/auto_email/auto_email.xlsx"
email_address = "phongnthai@gmail.com"
read_excel_and_send_email(excel_file_path, email_address)
```

```python
import win32com.client as win32
import openpyxl
import os

def read_excel_and_send_email(excel_file_path, email_address):
    # Load the Excel workbook
    wb = openpyxl.load_workbook(excel_file_path)
    ws = wb['Sheet1']

    # Initialize variables to store table data
    table_data = []

    # Iterate through rows and columns to extract table data
    for row in ws.iter_rows(min_row=1, max_row=50, max_col=3, values_only=True):
        if row[0] not in (None, 0):
            table_data.append(row)

    # Close the workbook
    wb.close()

    # Construct HTML table from the table data
    html_table = "<table border='1'>"
    for row in table_data:
        html_table += "<tr>"
        for cell in row:
            html_table += f"<td>{cell}</td>"
        html_table += "</tr>"
    html_table += "</table>"

    # Send email with HTML table in the email body
    outlook = win32.Dispatch('outlook.application')
    mail = outlook.CreateItem(0)
    mail.To = email_address
    mail.Subject = "Data from Excel Sheet1"
    mail.HTMLBody = f"<p>Here is the data from Excel Sheet1:</p>{html_table}"

    # Send the email
    mail.Send()

    print("Email sent successfully.")

# Example usage
excel_file_path = "C:/Users/phong/OneDrive/Desktop/auto_email/auto_email.xlsx"
email_address = "phongnthai@gmail.com"
read_excel_and_send_email(excel_file_path, email_address)
```
