```python
import os
import pandas as pd
import yfinance as yf
from datetime import datetime

# Assuming df is already defined and contains a column 'Symbol'
symbols = df['Symbol'].tolist()

# Filter out non-string symbols
symbols = [symbol for symbol in symbols if isinstance(symbol, str)]

end_date = datetime.today().strftime('%Y-%m-%d')
start_date = '1975-01-01'

df_append = []

for ticker in symbols:  # Iterating over symbols list
    df_ticker = yf.download(ticker, start=start_date, end=end_date, progress=False)
    df_ticker['ticker'] = ticker
    df_ticker.reset_index(drop=False, inplace=True)
    df_append.append(df_ticker)

    # Create directory if it doesn't exist
    directory = f'C:/Users/phong/OneDrive/Desktop/usstock/price'
    if not os.path.exists(directory):
        os.makedirs(directory)

    csv_file_path = f'{directory}/{ticker}.csv'
    df_ticker.to_csv(csv_file_path)

df_combined = pd.concat(df_append)
```
