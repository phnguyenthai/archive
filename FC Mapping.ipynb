```python
merged_df = pd.merge(df_expanded, df1, on=['edd', 'storagetype', 'fc_mapping'], how='left')
```

```python
merged_df.head()
```

```python
sorted_df = merged_df.sort_values(by=['edd', 'storagetype', 'fc_mapping', 'tbr'], ascending=[True, True, True, False])
```

```python
sorted_df['concat'] = sorted_df['edd'] + '_' + sorted_df['storagetype'] + '_' + sorted_df['fc_mapping']
```

```python
sorted_df
```

```python
# Define the allocation function
def allocate_capacity(group):
    available_capacity = group['available_capacity'].iloc[0]
    allocations = []
    for tbr in group['tbr']:
        if tbr <= available_capacity:
            allocations.append(tbr)
            available_capacity -= tbr
        else:
            allocations.append(0)
    return allocations

# Apply the allocation function to each group
sorted_df['allocation'] = sorted_df.groupby('concat').apply(lambda x: allocate_capacity(x)).explode().astype(int).values
```
