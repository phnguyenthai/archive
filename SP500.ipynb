```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import yfinance

raw_data = yfinance.download (tickers = "^GSPC", start = "1994-01-07", 
                              end = "2025-09-01", interval = "1d")
```

```python
# Reset index
raw_data.reset_index(drop=False, inplace=True)
df = raw_data[['Date','Close']]
```

```python
import pandas as pd

# Assuming df is your original DataFrame with columns Date and Close

# Create a copy of the DataFrame to avoid modifying the original DataFrame
df_copy = df.copy()

# Convert Date column to datetime
df_copy['Date'] = pd.to_datetime(df_copy['Date'])

# Extract year from Date column
df_copy['Year'] = df_copy['Date'].dt.year

# Find the maximum Date for each year
max_dates = df_copy.groupby('Year')['Date'].max().reset_index()

# Drop the 'Year' column from max_dates DataFrame
max_dates.drop('Year', axis=1, inplace=True)

# Merge with the original DataFrame to get corresponding Close values
df1 = pd.merge(max_dates, df_copy, on='Date', how='left')[['Date', 'Close']]

# Optional: Drop the 'Year' column if you don't need it anymore
# df1.drop('Year', axis=1, inplace=True)
```
