```python
# Path to the directory containing the CSV files
folder_path = 'C:/Users/phong/OneDrive/Desktop/usstock/price'

# List all files in the directory
files = os.listdir(folder_path)

# Filter out the CSV files
csv_files = [file for file in files if file.endswith('.csv')]

# Initialize an empty list to hold DataFrames
df_list = []

# Loop through the CSV files and read them into DataFrames
for csv_file in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    df = pd.read_csv(file_path)
    df_list.append(df)

# Concatenate all DataFrames into a single DataFrame
df  = pd.concat(df_list, ignore_index=True)
```

```python
price = df[['Date','Close','ticker']]
```

```python
price_df = price.rename(columns={'Date': 'date', 'Close': 'close'})
```

```python
price_df.head()
```

```python
price_df['date'] = pd.to_datetime(price_df['date'])

# Group by 'ticker' and resample to get the last date of each month
monthly_price = price_df.groupby(['ticker', pd.Grouper(key='date', freq='M')]).last().reset_index()
```

```python
monthly_price.shape
```

```python
# Shift the 'close' column by one month within each ticker group
monthly_price['next_month_price'] = monthly_price.groupby('ticker')['close'].shift(-1)
```

```python
# Shift the 'close' column by one month within each ticker group
monthly_price['last_month_price'] = monthly_price.groupby('ticker')['close'].shift(1)
```

```python
# Shift the 'close' column by 12 months within each ticker group
monthly_price['last_12m_price'] = monthly_price.groupby('ticker')['close'].shift(12)
```

```python
# Shift the 'close' column by 36 months within each ticker group
monthly_price['last_36m_price'] = monthly_price.groupby('ticker')['close'].shift(36)
```

```python
#max ever
monthly_price['max price ever'] = monthly_price.groupby('ticker')['close'].transform('max')
```

```python
# Assuming monthly_price is your DataFrame containing columns 'date', 'close', 'ticker'

# Ensure the 'date' column is of datetime type
monthly_price['date'] = pd.to_datetime(monthly_price['date'])

# Define a function to calculate the max price until the end of the previous year
def max_price_until_last_year(group):
    group = group.sort_values(by='date')
    max_prices = []
    for i, current_date in enumerate(group['date']):
        last_year_end = pd.Timestamp(year=current_date.year - 1, month=12, day=31)
        max_price = group[group['date'] <= last_year_end]['close'].max()
        max_prices.append(max_price)
    return pd.Series(max_prices, index=group.index)

# Apply the function to each group of ticker
monthly_price['max price ever until last year'] = monthly_price.groupby('ticker').apply(max_price_until_last_year).reset_index(level=0, drop=True)
```

```python
monthly_price.head()
```

```python
# Convert the 'date' column to datetime if it's not already
monthly_price['date'] = pd.to_datetime(monthly_price['date'])

# Group by 'ticker' and resample to get the last date of each quarter
quarterly_price = monthly_price.groupby(['ticker', pd.Grouper(key='date', freq='Q')]).apply(lambda x: x.loc[x['date'].idxmax()]).reset_index(drop=True)
```

```python
# Convert the 'date' column to datetime if it's not already
quarterly_price['date'] = pd.to_datetime(quarterly_price['date'])

# Create the 'quarter' column with the desired format
quarterly_price['quarter'] = quarterly_price['date'].dt.to_period('Q').astype(str).str.replace('Q', '-Q')
```

```python
output = pd.merge(ratio,quarterly_price,on = ['ticker','quarter'],how='left')
```

```python
con1 = output['pe'] >=5
con2 = output['roa'] >=0.25
output = output[con1 & con2]
```

```python
output
```

```python
output['pe_rank'] = output.groupby('quarter')['pe'].rank(ascending=True, method='dense')
output['roa_rank'] = output.groupby('quarter')['roa'].rank(ascending=False, method='dense')
output['total_rank']  = output['pe_rank'] + output['roa_rank']
output['final_rank'] = output.groupby('quarter')['total_rank'].rank(ascending=True, method='dense')
```

```python
output.head()
```

```python


# Ensure the 'quarter' column is of Period type for easier manipulation
output['quarter'] = pd.PeriodIndex(output['quarter'], freq='Q')

# Sort the DataFrame by ticker and quarter to ensure proper shifting
output = output.sort_values(by=['ticker', 'quarter'])

# Create the 'price same quarter next year' column by shifting 'close' within each ticker group by 4 quarters
output['price same quarter next year'] = output.groupby('ticker')['close'].shift(-4)

# Convert 'quarter' back to string for consistency if needed
output['quarter'] = output['quarter'].astype(str)
```

```python
output['return'] = output['price same quarter next year'] / output['close'] - 1
```

```python
# write df to csv file and open it.
csv_file_path = 'C:/Users/phong/OneDrive/Desktop/usstock/output.csv'
output.to_csv(csv_file_path, index=False)
# Open the CSV file
os.startfile(csv_file_path)
```
